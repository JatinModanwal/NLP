```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.txt',
                    sep='\t',names=["label","message"])
```

```python
messages
```

```python
import numpy as np
```

```python
from nltk.corpus import stopwords
import nltk
import re
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    reviews=re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    reviews=reviews.lower()
    reviews=reviews.split()
    
    reviews=[lemmatizer.lemmatize(word) for word in reviews]
    reviews=' '.join(reviews)
    corpus.append(reviews)
```

```python
# finding missing value
[[i,j,k]for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
```

```python
corpus
```

```python
# you use this code instead limmatizer
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
### lets train Word2vec from scratch
model = gensim.models.Word2Vec(words, window = 5, min_count = 2)
```

```python
# Get All the vocabulary
model.wv.index_to_key
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    # sent = [word for word in doc if word in model.wv.index_to_key]
    # print(sent)
    
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis = 0)
             # or [np.zeros(len(model.wv.index_to_key))], axis=0)
```

```python
#!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
# apply  for the entire sentences
X = []
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
# This is the final independent features
df_list = []

for i in range(len(X)):
    df_list.append(pd.DataFrame(np.array(X[i]).reshape(1, -1)))

df = pd.concat(df_list, ignore_index=True)
```

```python
## Dependent Feature
## Output Features

y= messages[list(map(lambda x: len(x)>0, corpus))]
y=pd.get_dummies(y['label'])
y=y.iloc[:,0].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
df['Output'] = y
```

```python
## train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.20)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=50, random_state=42)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_pred,y_test))
```
